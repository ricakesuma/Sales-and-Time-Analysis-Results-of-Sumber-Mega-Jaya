###Menghilangkan Duplicate Data

```python
# Menampilkan jumlah data duplikat
print(df.duplicated().sum())
```

```python
# Menghapus data duplikat sementara untuk memeriksa keseluruhan DataFrame
duplicates_removed_data = df.drop_duplicates()
```

```python
# Memeriksa apakah ada data duplikat dalam DataFrame
print(duplicates_removed_data.duplicated().any())
```

```python
# List kolom yang berisi tanggal (date)
date_columns = ['Order Date', 'Payment Date', 'Shipping Label Printed Date', 'Ready to Ship Date',
                'Shipped Date', 'Delivered Date', 'Completed Date', 'Cancelled Date']

# Ubah tipe data kolom-kolom tersebut menjadi datetime
for column in date_columns:
    df[column] = pd.to_datetime(df[column])

# Tampilkan hasil perubahan tipe data
print(df.dtypes)
```

```python
# Menampilkan jumlah nilai yang hilang pada setiap kolom
print(df.isnull().sum())
```

```python
df.dropna(subset=['Store'], inplace=True)
```

```python
# Mengisi missing values pada kolom 'Payment Type' dengan nilai 'Not Specified'
df['Payment Type'].fillna('Not Specified', inplace=True)

# Mengisi missing values pada kolom 'Payment Date' dengan nilai 'Unknown'
df['Payment Date'].fillna('Unknown', inplace=True)

# Mengisi missing values pada kolom 'Shipping Label Printed Date' dengan nilai 'Not Printed'
df['Shipping Label Printed Date'].fillna('Not Printed', inplace=True)
```
