# Jenis Pembayaran

### 1. Bank

```python
# BCA
df['Payment Type'] = df['Payment Type'].replace(['BCA Virtual Account','BCA_VA','KlikBCA','BCA KlikPay','BCA OneKlik'], 'BCA')
```

```python
# BNI
df['Payment Type'] = df['Payment Type'].replace(['BNI Virtual Account','BNI_VA'], 'BNI')
```

```python
# BRI
df['Payment Type'] = df['Payment Type'].replace(['BRIVA','Direct Debit BRI','BRI Ceria','bri_e_pay','BRILink','BRImo'], 'BRI')
```

```python
# Mandiri
df['Payment Type'] = df['Payment Type'].replace(['Mandiri Virtual Account','MANDIRIMANDIRI_VA','Direct Debit Mandiri','DOKU_MANDIRI_IPP'], 'Mandiri')
```

```python
# BSI
df['Payment Type'] = df['Payment Type'].replace(['Bank Syariah Indonesia'], 'BSI')
```

```python
# CIMB
df['Payment Type'] = df['Payment Type'].replace(['CIMB Virtual Account','OCTO Cash by CIMB Niaga'], 'CIMB')
```

```python
# Danamon
df['Payment Type'] = df['Payment Type'].replace(['Danamon Virtual Account'], 'Danamon')
```

### 2. Minimarket

```python
# Indomaret
df['Payment Type'] = df['Payment Type'].replace(['Indomaret / Ceriamart','INDOMARET_OTC'], 'Indomaret')
```

```python
# Alfamrt
df['Payment Type'] = df['Payment Type'].replace(['Alfamart / Alfamidi / Lawson / Dan+Dan','Alfamart / Alfamidi / Dan+Dan','ALFAMART_OTC'], 'Alfamart')
```

### 3. Fintech

```python
# Shopee Pay
df['Payment Type'] = df['Payment Type'].replace(['SPayLater','ShopeePay balance'], 'ShopeePay')
```

```python
# GoPay
df['Payment Type'] = df['Payment Type'].replace(['GoPay Coins','GoPayLater Cicil','GoPayLater'], 'GoPay')
```

### 4. Ecommerce

```python
# Tokopedia
df['Payment Type'] = df['Payment Type'].replace(['Saldo Tokopedia','Bebas Bayar dari Tokopedia'], 'Tokopedia')
```

```python
# Shopee
df['Payment Type'] = df['Payment Type'].replace(['SPayLater','ShopeePay balance'], 'ShopeePay')
```

### Exporting Dataset

```python
df = pd.DataFrame(df)
df.to_excel('filename.xlsx', sheet_name = 'New_sheet')
```
